##### Q.6

```python
def custom_float(var):
    try:
        return float(var)
    except ValueError:
        pass
    return np.nan
```

```python
housing['NUM_BATH'] = housing['NUM_BATH'].apply(lambda x: x if not isinstance(x, str) else custom_float(x)).fillna(method='ffill')
housing
```
